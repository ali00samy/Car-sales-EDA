```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
car_sales_df = pd.read_excel('Car_sales.xlsx')

car_sales_df.head(5)
```

```python
car_sales_df['Latest_Launch'] = pd.to_datetime(car_sales_df['Latest_Launch'])
```

```python
car_sales_df['Launch_Year'] = car_sales_df['Latest_Launch'].dt.year
```

```python
sales_trends = car_sales_df.groupby('Launch_Year')['Sales_in_thousands'].sum().reset_index()

sales_trends

plt.figure(figsize=(10,6))
plt.plot(sales_trends['Launch_Year'], sales_trends['Sales_in_thousands'],marker ='o', color='b' )
plt.grid(True)
plt.title('Sales Trend Over Years', fontdict={'fontweight' : 'bold'} )
plt.xlabel('Launch Year')
plt.ylabel('Sales in Thousands')
plt.xticks([2008,2009,2010,2011,2012])

plt.show()
```

```python
top_manufacturers = car_sales_df.groupby('Manufacturer')['Sales_in_thousands'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
bars = top_manufacturers.plot(kind = 'bar', color = 'c')
plt.title('Top 10 Manufacturers by Sales')
plt.xlabel('Manufacturer')
plt.ylabel('Total Sales in Thousands')
plt.xticks(rotation=45)

for bar in bars.patches:
    bars.text(
        bar.get_x() + bar.get_width() / 2,  
        bar.get_height() / 2,               
        f'{bar.get_height():,.0f}',         
        ha='center', va='center', color='white', fontsize=10
    )

plt.show()
```

```python
top_models = car_sales_df.groupby('Model')['Sales_in_thousands'].sum().sort_values(ascending=False).head(10)


plt.figure(figsize=(12,6))
bars = top_models.plot(kind = 'bar' , color = 'm')
plt.title('Top 10 models by sales')
plt.ylabel('Total Sales in Thousands')
plt.xlabel('Model')
plt.xticks(rotation=45)

for bar in bars.patches:
    bars.text(
        bar.get_x() + bar.get_width() / 2, 
        bar.get_height() / 2,               
        f'{bar.get_height():,.0f}',         
        ha='center', va='center', color='white', fontsize=10
    )

plt.show()
```

```python
vehicle_type_sales = car_sales_df.groupby('Vehicle_type')['Sales_in_thousands'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
bars = vehicle_type_sales.plot(kind='bar', color='skyblue')
plt.title('Sales by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Total Sales in Thousands')
plt.xticks(rotation=45)

# Add labels to each bar
for bar in bars.patches:
    bars.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() / 2,
        f'{bar.get_height():,.0f}',
        ha='center', va='center', color='black', fontsize=10
    )

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(car_sales_df['Price_in_thousands'],car_sales_df['Sales_in_thousands'], alpha=0.8 ,color='teal')
plt.title('Sales vs Price')
plt.xlabel('Price in Thousands')
plt.ylabel('Sales in Thousands')
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(car_sales_df['Fuel_efficiency'], car_sales_df['Sales_in_thousands'], alpha=0.6, color='green')
plt.title('Fuel Efficiency vs. Sales')
plt.xlabel('Fuel Efficiency (mpg)')
plt.ylabel('Sales in Thousands')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(car_sales_df['Horsepower'], car_sales_df['Sales_in_thousands'], alpha=0.6, color='purple')
plt.title('Horsepower vs. Sales')
plt.xlabel('Horsepower')
plt.ylabel('Sales in Thousands')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
car_sales_df['Horsepower'].plot(kind='hist', bins=20, color='orange', edgecolor='black')
plt.title('Distribution of Horsepower in Car Models')
plt.xlabel('Horsepower')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
correlation_matrix = car_sales_df[['Price_in_thousands', 'Horsepower', 'Fuel_efficiency', 'Sales_in_thousands']].corr()
```

```python
plt.figure(figsize=(8, 4))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, linewidths=0.5)

plt.title('Correlation Heatmap')
plt.xlabel('Features')
plt.ylabel('Features')
plt.xticks(fontsize=8,)
plt.yticks(fontsize=8)
plt.show()
```
